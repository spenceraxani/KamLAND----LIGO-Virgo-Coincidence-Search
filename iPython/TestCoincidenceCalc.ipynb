```python
def closest(event_list, k): 
    # Given the event_list, return the closest number to k.
    return event_list[min(range(len(event_list)), key = lambda i: abs(event_list[i]-k))] 

y1 = []
y2 = []
x = []
for i in np.arange(0.1,2,0.1):
    rA = 10
    nA = 1000

    rB = 20
    nB = 1000
    T = np.round(i,2)
    #print(T)
    f1 = np.random.exponential(scale=rA, size=nA)
    f1 = np.cumsum(f1)

    f2 = np.random.exponential(scale=rB, size=nB)
    f2 = np.cumsum(f2)


    n_coincidences = 0

    for i in range(len(f1)):
        #print(np.abs(closest(f2,f1[i])-f1[i]))
        if np.abs(closest(f2,f1[i])-f1[i]) < T:
            n_coincidences += 1 
    
    y1.append(n_coincidences/max(f1))
    y2.append(2*1/rA*1/rB*T)
    x.append(T)
    #print(n_coincidences/max(f1))
    #print(2*1/rA*1/rB*T)
```

```python

plt.plot(x,y1,label = 'Simulation')
plt.plot(x,y2,label = 'Calculation')
plt.xlabel('Coincidence time [s]')
plt.ylabel('Coincidence rate [Hz]')
plt.legend()
```
