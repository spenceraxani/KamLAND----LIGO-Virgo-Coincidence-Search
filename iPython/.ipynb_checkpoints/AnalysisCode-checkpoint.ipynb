```python
N90_long = 2.365
N1 = 175
N2 = 38
NT = 5.98E31
ET = 0.9
ES = KLZFit.DetectionEfficiency
xs = KLZFit.Constants.CrossSection
Phi = []
for i in range(len(KLZFit.Constants.NeutrinoCenterEnergies)):
    E_exp = KLZFit.Constants.NeutrinoCenterEnergies[i]#-0.78
    integral = []
    for j in range(len(KLZFit.Constants.NeutrinoCenterEnergies)):
        E_vis = KLZFit.Constants.NeutrinoCenterEnergies[j]
        R = 1/(2*3.14*0.064/np.sqrt(E_exp))*np.exp(-(E_exp - E_vis)**2/(2*(0.064/np.sqrt(E_exp))**2))
        integral.append(ES[j]*R*xs[j])
    s = sum(integral)
    #s = ES[i] *xs[i]
    Phi.append(N90_long/(N1*NT*ET*s))
    
```

```python
for confidenceLevel in [0.9]:
    lower_limit = []
    upper_limit = []
    for b in [0.18]: # Specify the range of backgruond events.
        print(b)
        fig, ax1 = plt.subplots()
        n_low = []      # Lower limit on Number observed.
        n_high = []     # Upper limit on Number observed.
        u = np.arange(0, 10, 0.02) # Signal mean. Hypotheses.
        n = np.arange(0,  20,1) # Number of observed events.

        for hyp in u: # loop through the different amounts of signal hypotheses. 
            Prob = []
            R    = []
            for n_obs in n: # loop through the different number of observations.
                hyp_LLH = scipy.stats.distributions.poisson.pmf(n_obs, b+hyp) # Calc poisson probability. Expectation is signal + background = b+hyp. Observed = n_obs.
                Prob.append(hyp_LLH)
                LLH = []
                for u_test in u:
                    LLH.append(scipy.stats.distributions.poisson.pmf(n_obs, b+u_test))
                max_LLH  = max(LLH) # given n_obs, wat is the best u_test. It's the one with the max LLH.
                R.append(hyp_LLH/max_LLH) # R factor from FC paper.
            R = np.asarray(R)
            sorted_R, sorted_n, sorted_P = (list(t) for t in zip(*sorted(zip(R, n, Prob),reverse = True))) # sort lists.

            sum_P = 0
            accepted_values = []

            for k in range(len(sorted_R)):
                sum_P += sorted_P[k]
                if sum_P < confidenceLevel:
                    accepted_values.append(sorted_n[k])
                else:
                    accepted_values.append(sorted_n[k])
                    break
            
            #print(hyp,len(accepted_values),min(accepted_values),max(accepted_values))
            n_low.append(min(accepted_values))
            if len(accepted_values) == 1:
                n_high.append(1)
            else:
                n_high.append(max(accepted_values))
            

        n_low = np.asarray(n_low)
        n_high = np.asarray(n_high)
        
        u_allowed_values = []
        
        outdict = {}
        
        for nobserved in n:
            outdict[nobserved] = []
            for k in range(len(u)):
                if np.logical_and(nobserved >= n_low[k], n_high[k]> nobserved):
                    outdict[nobserved].append(u[k])
                
                #if min(n_low[k])== max(n_high[k]):
                #    outdict[nobserved].append(u[k])

        for nobserved in n:
            try:
                outdict[nobserved] = [min(outdict[nobserved]),max(outdict[nobserved])]
            except:
                pass

        for k in range(len(n_low)):
            plt.plot([n_low[k],n_high[k]],[u[k],u[k]],'k',linewidth = 2)

        for nobserved in n:
            try:
                print([outdict[nobserved][0],outdict[nobserved][1]])
                plt.plot([nobserved+0.5,nobserved+0.5],[outdict[nobserved][0],outdict[nobserved][1]],color = 'r',linewidth = 5)
                #outdict[nobserved] = [min(outdict[nobserved]),max(outdict[nobserved])]
            except:
                pass



        plt.xlabel('N Observed')
        plt.ylabel('Mean Signal (u)')
        plt.grid(True)
        plt.axis([0,15,0,15])
        plt.show()
```
